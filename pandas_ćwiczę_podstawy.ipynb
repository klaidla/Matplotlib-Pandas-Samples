```python
d = pd.read_csv('ML-datasets/diabetes.csv')
d.head()
```

```python
d.Class.unique()
```

```python
d['Class_num'] = d.apply(lambda x: 1 if x.Class == 'positive' else -1, axis=1)
```

```python
d.head()
```

```python
d['my_factor'] = d.apply(lambda x: x['Triceps skin fold thickness']+x['Plasma glucose concentration']
                         if not x['Triceps skin fold thickness'] is None else x['Plasma glucose concentration'],
                         axis=1
                         )
d.head()
```

```python
import numpy as np
```

```python
d['Number of times pregnant']==0
```

```python
d.loc[d['Number of times pregnant']==0, 'Diastolic blood pressure']
```

```python
d.loc[d['Number of times pregnant']==0, 'Number of times pregnant'] = np.nan
# loca, gdy naraz filtruję i wybieram konkretną kolumnę
d.head()
```

```python
d['Number of times pregnant'].fillna(0)
```

```python
d.head()
```

```python
# d['Number of times pregnant'] = d['Number of times pregnant'].fillna(0)

d['Number of times pregnant'].fillna(0, inplace=True)
```

```python
d.head()
```

```python
d.loc[d['Number of times pregnant']==0, 'Number of times pregnant'] = np.nan
d.head()
```

```python
mean_times_pregnant = d['Number of times pregnant'].mean().astype(int)
mean_times_pregnant
```

```python
d['Number of times pregnant'].fillna(mean_times_pregnant, inplace=True)
d.head()
```

```python
mode_times_pregnant = d['Number of times pregnant'].mode()
d['Number of times pregnant'].fillna(mode_times_pregnant, inplace=True)
d.head()
```

```python
d.loc[d['Number of times pregnant']==0, 'Number of times pregnant'] = np.nan

d['Number of times pregnant'].interpolate(inplace=True) 
d.head()
```

```python
d.describe()
```

```python
d['2-Hour serum insulin'] = (d['2-Hour serum insulin'] - d['2-Hour serum insulin'].mean())/d['2-Hour serum insulin'].std()
```

```python
d.describe()
```

```python
# d = d.sort_values(by='Diastolic blood pressure')
d.sort_values(by='Diastolic blood pressure', inplace=True,
              na_position='first' #'last'
              )
d
```
