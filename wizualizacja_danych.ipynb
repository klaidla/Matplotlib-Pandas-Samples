```python
def plot_timeseries(axis, x, y, c, xlabel, ylabel):
  axis.plot(x, y, color=c)
  axis.set_xlabel(xlabel)
  axis.set_ylabel(ylabel, color=c)
  axis.tick_params('y', colors=c)
```

```python
def groupby_column(dataframe, column):
  groupby_obj = dataframe.groupby(column)
  return {x: groupby_obj.get_group(x) for x in groupby_obj.groups} 
```

```python
# to samo, co groupby_column, wersja 2
def groupby_column2(dataframe, column):
  my_dict = {}
  for x in dataframe[column].unique():
    subset = dataframe[dataframe[column] == x]
    my_dict[x] = subset
  return my_dict
```

```python
def analyze_medals(medal_df):
  countries = groupby_column(medal_df, 'NOC')
  ret = {}
  for country, df in countries.items():
    counts = df["Medal"].value_counts()
    gold = counts['Gold'] if 'Gold'in counts else 0 
    # if 'Gold' in counts:
    #   gold = counts['Gold']
    # else:
    #   gold = 0
    silver = counts['Silver'] if 'Silver' in counts else 0
    bronze = counts['Bronze'] if 'Bronze' in counts else 0
    ret[country] = (gold, silver, bronze)
  return ret
# analyze_medals(medals)
```

```python
c_df = pd.DataFrame.from_dict(analyze_medals(medals),
                              orient='index',
                              columns =['Gold', 'Silver', 'Bronze'])
c_df.head()
```

```python
new_medals = c_df[c_df.sum(axis=1) > 500]
new_medals.head()
```

```python
# zadanie - skumulowany wykres supkowy
# x - kod kraju, wysoko supka - liczba medali z podziaem na kolor medalu
fig, ax = plt.subplots()
ax.bar(new_medals.index, new_medals["Gold"], label="Gold")
ax.bar(new_medals.index, new_medals["Silver"],
       bottom=new_medals["Gold"], label="Silver")
ax.bar(new_medals.index, new_medals["Bronze"],
       bottom=new_medals["Gold"]+new_medals["Silver"], label="Bronze")
ax.xaxis.set_tick_params(rotation=90)
ax.set_ylabel('Medals')

ax.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

#define data
fig, ax = plt.subplots()
fig.subplots_adjust(0.1, 0.1, 1, 1)

data = [15, 25, 25, 30, 5]
labels = ['Group 1', 'Group 2', 'Group 3', 'Group 4', 'Group 5']

#define Seaborn color palette to use
colors = sns.color_palette('pastel')[0:5]
my_width = 0.3
ax.bar(labels, data, color = colors, width=my_width) #radiusdefault=1
for i, v in enumerate(data):
    ax.text(i-my_width/4, #pozycja x
            v+1, #pozycja y 
            str(v), #text
            color='blue', fontweight='bold', backgroundcolor='r')
plt.show()
# https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.text.html
```

```python
df= pd.DataFrame({'number': np.random.randint(1, 100, 10)})
my_bins = [1, 20, 40, 60, 80, 100]
df['bins'] = pd.cut(x=df['number'], bins=my_bins)
print(df)

plt.hist(df.number, bins=my_bins, edgecolor="k")
plt.xticks(my_bins)
 
# We can check the frequency of each bin
# print(df['bins'].unique())
```

```python
out= pd.cut(x=df['number'], bins=my_bins)
df['bins'] = out

df2 = df['bins'].value_counts().rename_axis('interval').reset_index(name='counts')
df2['i'] = df2.index
print(df2)

fig, ax = plt.subplots()
ax.bar(df2.i, df2.counts)
ax.set_xticks(df2.i)
ax.set_xticklabels(['({},{}]'.format(c.left, c.right) for c in df2.interval.cat.categories])
```
